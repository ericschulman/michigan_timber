```python
import os
import pandas as pd
```

```python
#get a list of the filenames
cwd = os.getcwd()
filelist = []
for root, directories, filenames in os.walk(cwd):
    for filename in filenames:
        if '.xls' in filename:
            path = str(os.path.join(root, filename))
            filelist = filelist + [path]
```

```python
print(len(pd.read_excel(filelist[0],skiprows=29)))
print((pd.read_excel(filelist[0],skiprows=29)).columns)
```

```python
years = [str(i) for i in range(2004,2013,1)]
filelist_short = []

for file in filelist:
    for year in years:
        if year in file:
            filelist_short = filelist_short + [file]
            
print(filelist_short)
```

```python
#read the files and then add them into a master file
result = pd.read_excel(filelist_short[0],skiprows=29)

for file in filelist_short[1:]:
    current = pd.read_excel(file,skiprows=29)
    #office_name = file[63:]
    #office_name = office_name[:office_name.find('/')]
    #current.insert(0, 'Location', office_name.capitalize())
    result = result.append(current,ignore_index=True,sort=True)
```

```python
result_final = result.loc[:, ~result.columns.str.contains('Unnamed: ')]
result_final = result_final.loc[:, ~result_final.columns.str.contains('two')]
result_final = result_final.loc[:, ~result_final.columns.str.contains('6')]
result_final.to_csv('mich_sales.csv')
```
