```python
import numpy as np
import pandas as pd
```

```python
mich_bids = pd.read_csv('mich_bids.csv',index_col=0)
```

```python
# filtering data for Baldwin Office
baldwin_office = pd.DataFrame(mich_bids[mich_bids['Bid Open Location'] == "Baldwin Office"])
baldwin_office
```

```python
# group by Sale #
pd.DataFrame(baldwin_office.groupby('Sale #'))
```

```python
# group by Sale #, Species
pd.DataFrame(baldwin_office.groupby('Species')['Sale #'].nunique())
```

```python
# group by Sale #, Product
pd.DataFrame(baldwin_office.groupby(['Product'])['Sale #'].nunique())
```

```python
# group by Sale #, Species, Product
pd.DataFrame(baldwin_office.groupby(['Species', 'Product'])['Sale #'].nunique())
```

```python
# group by Sale #, Product, Species
pd.DataFrame(baldwin_office.groupby(['Product', 'Species'])['Sale #'].nunique())
```

```python
# group by Date, Species
pd.DataFrame(baldwin_office.groupby('Species')['Bid Open Date'].nunique())
```

```python
# group by Date, Product
pd.DataFrame(baldwin_office.groupby('Product')['Bid Open Date'].nunique())
```

```python
# group by Date, Species, Product
pd.DataFrame(baldwin_office.groupby(['Species', 'Product'])['Bid Open Date'].nunique())
```

```python
# group by Date, Product, Species
pd.DataFrame(baldwin_office.groupby(['Product', 'Species'])['Bid Open Date'].nunique())
```
