Question A - What is the average bid if we exclude non-bids? i.e. The column "Bid/Non-bid species" is non-bid?

```python
mich_bids = pd.read_csv('mich_bids.csv')

mich_bids['bid_amount'] = mich_bids['Bid Per Unit'] * mich_bids['Estimated Volume']

mich_bids= mich_bids[mich_bids['Bid/Non-Bid Species'] == 'Bid']

mean = mich_bids['bid_amount'].mean()

mean
```

Question B - What is the average "Bid per Unit" for each value of "Species''? What is the average bid for each value of "Units"?

```python
mich_bids = pd.read_csv('mich_bids.csv')

sp_mean = mich_bids.groupby('Species').mean()

sp_mean['Bid Per Unit']
```

Question C - What is the average "Estimated Volume" for each value of "Species"? What about for each value of "Units"?

```python
sp_mean = mich_bids.groupby('Species').mean()

sp_mean['Estimated Volume']
```

Question D - Can non-bids win the auction? i.e. are there instances where "Highest" is not NaN and "Bid/Non-bid species" is a non-bid?

```python
mich_bid= mich_bids[mich_bids['Bid/Non-Bid Species'] == 'Bid']

bids_mean = mich_bid.groupby('Sale #').agg({ 
    'Bid/Non-Bid Species': ['count']})

mich_non_bids= mich_bids[mich_bids['Bid/Non-Bid Species'] == 'Non-Bid']

non_bids_mean = mich_non_bids.groupby('Sale #').agg({ 
    'Bid/Non-Bid Species': ['count']})


print('bid mean :', bids_mean.mean())

print('non-bid mean :', non_bids_mean.mean())

```

Question E - What is the average number of bids in an auction? What is the average number of non-bids in each auction?

```python
Cords = []
for row in mich_bids['Units']:
    if row == 'Cords':
        Cords.append(1)
    else:
        Cords.append(0)
        
MBF = []
for row in mich_bids['Units']:
    if row == 'MBF':
        MBF.append(1)
    else:
        MBF.append(0)

Acres = []
for row in mich_bids['Units']:
    if row == 'Acres':
        Acres.append(1)
    else:
        Acres.append(0)
        
mich_bids['Cords'] = Cords
mich_bids['MBF'] = MBF
mich_bids['Acres'] = Acres

mich_bids

Unit_count = mich_bids.groupby('Species').agg({ 
    'Cords': ['sum'],
    'MBF': ['sum'],
    'Acres': ['sum']
})

Unit_count
```
